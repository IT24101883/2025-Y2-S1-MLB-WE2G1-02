```python
from google.colab import drive
drive.mount('/content/drive')
import os


image_dir = '/content/drive/MyDrive/Vehicles/Motorcycles'

resized_dir = '/content/drive/MyDrive/resized_dataset'

os.makedirs(resized_dir, exist_ok=True)

from PIL import Image
import glob

image_extensions = ['*.jpg', '*.jpeg', '*.png', '*.bmp', '*.webp','*rgba']

for ext in image_extensions:
    for image_path in glob.glob(os.path.join(image_dir, ext)):
        try:
            img = Image.open(image_path)
            img = img.resize((224, 224))
            base_name = os.path.basename(image_path)
            img.save(os.path.join(resized_dir, base_name))

            print(f"Resized: {base_name}")
        except Exception as e:
            print(f"Error processing {image_path}: {e}")

```
